```
from transformers import pipeline
```

## Load Model

```
generator = pipeline('text-generation', model='gpt2')
```

## Input Ingredients

```
ingredients = 'rice, onion, tomato, spices'
prompt = f'Create a cooking recipe using these ingredients: {ingredients}. Include recipe name, ingredients list, and steps.'
```

## Generate Recipe

```
recipe = generator(prompt, max_length=200, num_return_sequences=1)
print(recipe[0]['generated_text'])
```
